```python
# Importing Required Library 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Startups.csv")   # Reading the Data .csv File
```

```python
from sklearn.preprocessing import LabelEncoder # feture Scaling 
label=LabelEncoder()
```

```python
dataset['State']=label.fit_transform(dataset['State'])
```

```python
x=dataset.iloc[:,:-1].values  # feture Extraction
y=dataset.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split #importing Requires maodel for  feture Extraction
```

```python
x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.25)   # feture Extraction
```

```python
from sklearn.linear_model import LinearRegression  # Importing Machine Learning LinearRegression Model from sklearn 
regressor=LinearRegression()  # creating model Object 
```

```python
regressor.fit(x_train,y_train) # fitting the liner Regression  model
```

```python
y_pred=regressor.predict(x_test) # Predicting model with the test data
```

```python
for mp, av in zip(y_pred, y_test):
  print ('Model Predicted Value ={}, Known Value = {}, difference = {}'.format(round(mp,2),round(av,2),round(abs(mp-av),2)))
```

```python
regressor.score(x_train,y_train)  # Cheacking the model Accuracy Score 
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse=mean_squared_error(y_pred,y_test)
mae=mean_absolute_error(y_pred,y_test)
print("Mean Squared Error={},\nMean Absour Error={}".format(mse,mae))
```
